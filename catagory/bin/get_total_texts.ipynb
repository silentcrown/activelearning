```python
class data():
    
    def __init__(self, data_path, undefined_data_path, undefined_text_path, text_path, reset, k):
        self.data = open(data_path, 'r').readlines()
        self.undefined_data = open(undefined_data_path, 'r').readlines()
        self.text_path = text_path
        self.undefined_text_path = undefined_text_path
        self.reset = reset
        self.train_indexes = []
        self.test_indexes = []
        self.k = k
        self.unlabeled = []
        self.labeled = []
        #self.sets = []
        
        
    def get_undefined_data(self, undefined_text_path):
        texts = []
        for i in range(1, len(self.undefined_data)):
            texts.append(self.undefined_data[i].split('%')[0])
        text_csv = pd.DataFrame({'text':texts})
        text_csv.to_csv(undefined_text_path, index = False, encoding = 'utf-8')
        
    def get_texts(self, text_path, reset = False):
        texts = []
        if reset == True:
            for i in range(1,len(self.data)):
                for key in self.data[i].split('||'):
                    texts.append(key)    
            text_csv = pd.DataFrame({'text':texts})
            text_csv.to_csv(text_path, index = False, encoding = 'utf-8')
    
    def split_train_and_test(self):
        self.get_texts(self.text_path, True)
        data = pd.read_csv(self.text_path)
        indexes = data.index.tolist()
        random.shuffle(indexes)
        self.train_indexes, self.test_indexes = indexes[:int(0.8 * len(indexes))], indexes[int(0.8 * len(indexes)) :]
        train, test = data.loc[self.train_indexes,'text'], data.loc[self.test_indexes,'text']
        self.get_undefined_data(self.undefined_text_path)
        undefined = list(pd.read_csv(self.undefined_text_path)['text'])
        return undefined,list(train), list(test)
    
    def get_labeled_and_unlabeled(self):
        undefined, train, test = self.split_train_and_test()
        new_train, new_undefined = [(train[i],'train',i) for i in range(len(train))], [(undefined[i],'undefined',i) for i in range(len(undefined)) if type(undefined[i])  != float and undefined[i].strip() != '' and '提醒' not in undefined[i]]
        new_train_indexes = range(len(new_train))
        random.shuffle(new_train_indexes)
        new_train_labeled, new_train_unlabeled = [new_train[i] for i in new_train_indexes[:1000]], [new_train[i] for i in new_train_indexes[1000:]]
        new_unlabeled = new_train_unlabeled + new_undefined
        #new_unlabeled_indexes = range(len(new_unlabeled))
        
        #random.shuffle(new_unlabeled_indexes)
        self.unlabeled = new_unlabeled
        random.shuffle(self.unlabeled)
        self.labeled = new_train_labeled
        random.shuffle(self.labeled)
        #self.sets = [new_unlabeled_indexes[i:i + self.k] for i in xrange(0, len(new_unlabeled_indexes), self.k)]        
```

```python
class word():    
    def __init__(self):
        self.word_to_vec = {}
        self.dictionary = corpora.dictionary.Dictionary()
        self.dic = self.dictionary.load('../util/new/laiye/corpus.dict')
        self.tfidf = models.TfidfModel.load('../util/new/laiye/corpus.tfidf_model')  
        
    def remove_punctuation(self, line):
        rule = re.compile(ur"[^a-zA-Z0-9\u4e00-\u9fa5]")
        line = rule.sub('',line)
        return line        

    def stopwordslist(self, filepath):  
        stopwords = [line.strip() for line in open(filepath, 'r').readlines()]  
        return stopwords  


    # 对句子进行分词  
    def jieba_cut(self, sentence):  
        sentence_seged = jieba.cut(sentence.strip())  
        stopwords = stopwordslist('../util/new/laiye/stopwords.txt')  # 这里加载停用词的路径  
        outstr = ''  
        for word in sentence_seged:  
            if word.encode('utf-8') not in stopwords:  
                if word.encode('utf-8') != '\t' and word != '\t':  
                    outstr += word
                    outstr += " "  
        return outstr.strip().split(' ')

    def get_word2vec(self):
        with open('../util/new/laiye/w2v_sgns_win1_d80.kv') as f:
            data = [x.split(' ') for x in f.readlines()[1:]]
            words = [d[0] for d in data]
            vecs = np.array([d[1: -1] for d in data], dtype= 'float64')
            for i in range(len(words)):
                word = words[i]
                self.word_to_vec[word] = vecs[i]

    def get_tf_idf_of_query(self, query, dic, tfidf):      
        vec_bow = dic.doc2bow(query)
        vec_tfidf = tfidf[vec_bow]
        tp = [0.0] * len(query)
        ids = [tid[0] for tid in vec_tfidf]
        flags = [0] * len(query)
        count = 0.00001
        for j in range(len(query)):
            for i in range(len(ids)):
                if self.dic[ids[i]] == query[j]:
                    tp[j] = vec_tfidf[i][1]
                    flags[j] = 1
                    count += 1
                    break
        sums = sum(tp)
        #print ','.join(query)
        for i in range(len(flags)):
            if flags[i] == 1:
                continue
            tp[i] = 1.0 * sums / count
                    
        # apply l1-norm to tfidf value
        #tfidf = matutils.unitvec(vec_tfidf, norm = 'l1')   
        return tp

    def get_vectors_of_data_cut(self, data):
        vecs = []
        for i in range(len(data)):
            vec = []
            d_line = data[i].split(' ')
            tfidfs = self.get_tf_idf_of_query(d_line, self.dic, self.tfidf)
            s = sum(tfidfs)
            for j in range(len(d_line)):
                if d_line[j].encode('utf-8') in self.word_to_vec:
                    if s == 0:
                        vec.append([0.0] * len(self.word_to_vec['家']))
                    else:
                        vec.append((tfidfs[j] / s) * self.word_to_vec[d_line[j].encode('utf-8')])
            if len(vec) == 0:
                vec.append([0.0] * len(self.word_to_vec['家']))
            vecs.append(np.sum(np.array(vec), axis = 0))
        return np.array(vecs)    
    
    
```

```python
class active_learning():
    
    def __init__(self, data_util, word_util, k):
        self.k = k
        self.data_util = data_util
        self.word_util = word_util
        self.sets = []
        self.remove = []

    def get_random_sets(self, unlabeled):
        #self.get_unlabeled_index()
        random.shuffle(unlabeled)
        self.sets = [unlabeled[i:i + self.k] for i in xrange(0, len(unlabeled), self.k)]
        print len(self.sets)
        
    def get_cur_set_texts(self, cur_set):
        add_texts = [cur_set[i][0] for i in range(len(cur_set))]
        return add_texts
        
    def get_labeled_and_unlabeled_texts(self):
        unlabeled_texts = [self.data_util.unlabeled[i][0] for i in range(len(self.data_util.unlabeled))]
        unlabeled_cat = [self.data_util.unlabeled[i][1] for i in range(len(self.data_util.unlabeled))]
        labeled_texts = [self.data_util.labeled[i][0] for i in range(len(self.data_util.labeled))]
        return labeled_texts, unlabeled_texts, unlabeled_cat
    
    def get_vectors_of_data(self, texts):
        cuts, vecs = [], []
        for i in range(len(texts)):
            if type(texts[i]) != str:
                self.remove.append(i)
                continue
            if '提醒' in texts[i]:
                self.remove.append(i) 
                continue
            cuts.append(FasttextClassifier.fasttext_tokenize(self.word_util.remove_punctuation(texts[i].decode('utf-8'))))
        vecs = self.word_util.get_vectors_of_data_cut(cuts)  
        return vecs
                        
                        
    #def get_scores_of_texts():

    def cosSimilar(self, inA, inB):
        inA = np.mat(inA)
        inB = np.mat(inB)
        num = float(inA*inB.T)
        denom = np.linalg.norm(inA)*np.linalg.norm(inB)
        return 0.5+0.5*(num/denom)

    def compute_sim_by_word2vec_avr(self, a ,b):
        time_start=time.time()
        data_a = a
        data_b = b
        sims = 1 + np.dot(data_a,data_b.T)/(np.linalg.norm(data_a)*(np.linalg.norm(data_b)))
        time_end=time.time()
        print('time cost',time_end-time_start,'s')
        return sims 
    
    def get_f_score(self):
        self.get_random_sets(self.data_util.unlabeled)
        labeled_texts, unlabeled_texts, unlabeled_cat = self.get_labeled_and_unlabeled_texts()
        xl = self.get_vectors_of_data(labeled_texts) 
        xi = self.get_vectors_of_data(unlabeled_texts) 
        base_s, f_max = None, None
        print 'length of sets is : '+ str(len(self.sets))
        b = self.compute_sim_by_word2vec_avr(xi, xl)
        for i in range(len(self.sets)):
            print 'set ' + str(i)
            if len(self.sets[i]) < self.k:
                continue
            cur_set = self.sets[i]
            self.remove = []
            xs = self.get_vectors_of_data(self.get_cur_set_texts(cur_set)) 
            xa = np.concatenate((xs, xl), axis=0)
            a = self.compute_sim_by_word2vec_avr(xi, xa)
            #b = self.compute_sim_by_word2vec_avr(xi, xb)
            f = np.sum(np.max(a, axis = 1) - np.max(b, axis = 1))
            if f > f_max:
                f_max = f
                base_s = cur_set
            print 'current f is :' + str(f) + ',    max:' + str(f_max)
        return cur_set        
```
